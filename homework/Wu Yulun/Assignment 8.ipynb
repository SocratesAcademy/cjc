# 构建networkx的网络对象g

```python
G = nx.DiGraph()
n = 0
with open ('/Users/Wuyulunbi/Desktop/www.dat.gz.txt') as f:
    for line in f:
        n += 1
        if n % 10**4 == 0:
            flushPrint(n)
        x, y = line.rstrip().split(' ')
        G.add_edge(x,y)
```

```python
dc = nx.out_degree_centrality(G)
fig = plt.figure(figsize=(15, 4),facecolor='white')
ax = plt.subplot(1, 3, 1)
plt.hist(dc.values(), bins = 20)
plt.gca().set_xscale('log')
plt.gca().set_yscale('log')
plt.xlabel('$Out\,Degree \, Centrality$', fontsize = 20)
plt.ylabel('$Frequency, \, F$', fontsize = 20)
```
