```python
def flushPrint(variable):
    if variable %10^2 == 0:
        sys.stdout.write('\r')
        sys.stdout.write('%s' % variable)
        sys.stdout.flush()
```

```python
G = nx.DiGraph()
n = 0
with open ('/Users/dustin/Desktop/www.dat.gz.txt') as f:
    for line in f:
        n += 1
        if n % 10**4 == 0:
            flushPrint(n)
        x, y = line.rstrip().split(' ')
        G.add_edge(x,y)
```

## 绘制www网络的出度分布、入度分布

```python
dc = nx.in_degree_centrality(G)
fig = plt.figure(figsize=(15, 4),facecolor='white')
ax = plt.subplot
plt.hist(dc.values(), bins = 20)
plt.xlabel('$In\,Degree \, Centrality$', fontsize = 20)
plt.ylabel('$Frequency, \, F$', fontsize = 20)
```

```python
dc = nx.out_degree_centrality(G)
fig = plt.figure(figsize=(15, 4),facecolor='white')
ax = plt.subplot
plt.hist(dc.values(), bins = 20)
plt.xlabel('$Out\,Degree \, Centrality$', fontsize = 20)
plt.ylabel('$Frequency, \, F$', fontsize = 20)
```

# 使用BA模型生成节点数为m取值为2，N取值分别为10、100、1000、10000时，绘制平均路径长度d与节点数量的关系

```python
import networkx as nx
import matplotlib.pyplot as plt
BA= nx.random_graphs.barabasi_albert_graph(1000,2)  
pos = nx.spring_layout(BA)         
nx.draw(BA,pos,with_labels=False,node_size = 30)  
plt.show()
```

```python
Ns = [i*10 for i in [1, 10, 100, 1000]]
ds = []
for N in Ns:
    print N
    BA= nx.random_graphs.barabasi_albert_graph(N,2)
    d = nx.average_shortest_path_length(BA)
    ds.append(d)
```

```python
plt.plot(Ns, ds, 'r-o')
plt.xlabel('$N$', fontsize = 20)
plt.ylabel('$<d>$', fontsize = 20)
plt.xscale('log')
plt.show()
```
