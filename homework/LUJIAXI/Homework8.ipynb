## 构建networkx的网络对象g（提示：有向网络）
## 将www数据添加到g当中

```python
%matplotlib inline
import networkx as nx
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
n = 0
with open ('C:/Users/lenovo/Desktop/homework/www.dat/www.dat') as f:
    for line in f:
        n += 1
        x, y = line.rstrip().split(' ')
        G.add_edge(x,y)
```

## 计算网络中的节点数量和链接数量

```python
nx.info(G)
```

## 计算www网络的网络密度

```python
nx.density(G)
```

## 绘制www网络的出度分布、入度分布

```python
dc = nx.in_degree_centrality(G)
```

```python
fig = plt.figure(figsize=(15, 4),facecolor='white')
ax = plt.subplot(1, 3, 1)
plt.hist(dc.values(), bins = 20)
plt.xlabel('$In\,Degree \, Centrality$', fontsize = 20)
plt.ylabel('$Frequency, \, F$', fontsize = 20)
```

```python
dc2 = nx.out_degree_centrality(G)
```

```python
fig = plt.figure(figsize=(15, 4),facecolor='white')
ax = plt.subplot(1, 3, 1)
plt.hist(dc2.values(), bins = 20)
plt.xlabel('$Out\,Degree \, Centrality$', fontsize = 20)
plt.ylabel('$Frequency, \, F$', fontsize = 20)
```

## 使用BA模型生成节点数为N、幂指数为γγ的网络

```python
import networkx as nx
import matplotlib.pyplot as plt
BA= nx.random_graphs.barabasi_albert_graph(500,2)  
pos = nx.spring_layout(BA)          
nx.draw(BA,pos,with_labels=False,node_size = 30)  
plt.show()
```

## 计算平均路径长度d与节点数量的关系

```python
d = nx.average_shortest_path_length(BA)
d
```

```python
Ns = [i*10 for i in [1, 10, 100, 1000]]
ds = []
for N in Ns:
    print N
    BA= nx.random_graphs.barabasi_albert_graph(N,2)
    d = nx.average_shortest_path_length(BA)
    ds.append(d)
```

```python
plt.plot(Ns, ds, 'r-o')
plt.xlabel('$N$', fontsize = 20)
plt.ylabel('$<d>$', fontsize = 20)
plt.xscale('log')
plt.show()
```
