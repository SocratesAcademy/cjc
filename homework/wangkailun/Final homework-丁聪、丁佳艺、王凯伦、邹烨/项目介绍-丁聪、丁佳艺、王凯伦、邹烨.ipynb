## 计算新闻传播学——天涯论坛数据分析

## 项目摘要

我们选取天涯论坛关于PM2.5主题相关的数据，希望通过数据的挖掘，分析，和主题模型的建立，旨在发现天涯关于PM2.5讨论回帖中的数据特征和舆情分析。

## 项目思路

本研究在对天涯论坛的天涯杂谈板块关于PM2.5的主贴内容进行数据挖掘与清洗之后，第一步进行了数据呈现状况的可视化描述和相关性分析；第二步使用了tfid、textrank和停用词三种算法对天涯论坛的数据进行词云分析；第三步在对天涯回帖数据进行提取与停用词去除之后，基于LAD模型对回帖内容进行主题模型的构建；第四步通过对天涯回帖中用户回复帖子的字数来计算偏好值，然后计算任意两个用户的距离计算相似性向用户推荐帖子；第五步通过对天涯论坛回帖者网络关系节点的计算，找到网络关系数最多的节点，显示其网络，以上为本项目的研究思路。

## 引言

   2003年，PM2.5所引发的雾霾现象出现在中国。然而，直到2012年，伴随着雾霾的大规模爆发，这一环境问题才正式被人们所讨论和关注，进而上升为社会议题。最严重时，雾霾曾经席卷中国几乎所有的省份和城市，以至于随后的调查纪录片《穹顶之下》引爆了社会热点。
   
   彼时，最为老牌论坛社区——天涯，关于PM2.5的讨论也同样如火如荼。天涯网民对于PM2.5环境问题相关舆论的产生和发酵，扮演了何种角色？论坛中PM2.5的相关信息，又会呈现出哪些关键性的特征？
   
   基于此，我们希望通过挖掘天涯关于pm2.5相关的帖子，建立主题模型，分析回帖数据，探索天涯pm2.5相关的舆情的特点。

## 研究发现

### 数据分析部分
1.我们分别对原始数据，原始数据的对数进行了框须图的可视化呈现。能够很直观的感受到，在数据处理中，取对数是有效观测数据的方式。
2.框须图的特性：
能够真实直观地表现数据形状的本来面貌，由于判断异常值的标准以四分位数和四分位距为基础，四分位数具有一定的耐抗性，多达25%的数据可以变得任意远而不会很大地扰动四分位数，所以异常值不能对这个标准施加影响，因此，框须图识别异常值的结果比较客观。
3.利用框须图对天涯论坛关于pm2.5主题帖的点击.回复.关注 以及粉丝数进行分析，结果如下：
  总体来看：点击.回复.关注 以及粉丝数都存在异常值。其中，点击量总体数值大于回复量，粉丝量总体大于关注者。从图形可知，点击数量的正态分布优于回复帖数，粉丝数正态分布优于回复数。
  为了更好分析了解数据，对点击数，回复数以及粉丝数进行柱状分布图分析。
  
  我们总共抓取了193个天涯论坛关于pm2.5的主题帖，分析结果如下：
1.结合框须图分析，可知点击数呈现正态分布，回帖数量则呈现非正态分布；粉丝数分布优于关注者数量分布；
2.点击数量最主要分布在0～5000之间，从图形显示，大概在90%左右。点击数的最大值为31056，最小值为7；
3.回帖数量最主要分布在0～500，从图形显示，大概在90%左右。回帖数最大值为237，最小则为0。

通过对数据的处理，进行reply ~ click + followed_numde1相关性回归分析，log之后Regression Results中的r-square值显示为0.548;Prob值为2.14e-32,小于0.05；所以reply ~ click + followed_num数据的线性相关性不显著。
2.在控制变量之后，可以看出reply与click的具有相对较好的相关性。

### 文本分析部分

通过“基于tfidf过滤”、“基于TextRank算法”和“基于停用词过滤”三种不同的词云分析发现，空气、口罩、污染、城市、中国、北京、颗粒物、环境、严重、健康等词是参与者在“PM2.5”的议题下讨论的热点。比较三种方法，“基于tfidf过滤”的词云对停用词的处理最差，词云图中含有大量的无用词如“我们、一个、还是、可以”等；“基于TextRank算法”的词云对停用词的处理最好，分析的结果也与另外两种方法有明显的差异。

将所有回复贴分类合并后，再基于 TextRank 算法提取关键词并绘图，可以发现每年关于PM2.5的讨论重点有所差异，2012年“口罩”、“空气质量”、“污染”是参与者最爱提及的词汇，到了2013年，“空气”一次的热度提升，热议词变成了“空气”、“口罩”、“污染”，2014年则是“空气”、“口罩”、“污染”、“净化器”，虽然2013年也出现了“净化器”，但比重显然比不上2014年。2015年，“供暖”成为大家在讨论PM2.5时提及频率较高的词汇，热议词变成了“口罩”、“供暖”、“污染”和“空气”，“净化器”的讨论热度下降，极少被提及。然而到了2016年，“净化器”一次的讨论热度再次上升，成为2016年关于PM2.5议题的第二关键词，“空气”和“环境”仍是大家讨论的重点。

通过词向量的时间序列我们可以发现，“口罩”、“净化器”两个词呈负相关，当“口罩”的重要性上升时，“净化器”一词的重要性就会下降，反之亦然。2012年至2015年，“口罩”的重要性一直高于“净化器”，但在2016年出现了反转，“口罩”的重要性急剧下降，“净化器”的重要性急剧上升。再联系现实生活，口罩和净化器是人们应对环境污染和雾霾天气采取的两种措施，往年虽然有人提到了净化器，但讨论热度一直不高，而在2016年，净化器取代了口罩成为了人们关注的热点，其原因可能是科学技术的进步，使得净化器更受大众青睐。

“空气”、“口罩”和“污染”三个词一直是历年来的热议词，其中以“空气”一词最为突出，分值一直居高不下，而“口罩”和“污染”的重要性在2016年有所下降。

### 主题模型

通过文本处理，共取出20个主题，总计200种概念；第4组，9组，13组，19组，主题权重整体较高，分别概括为：PM2.5新闻（4组）；政治（9组）；环境（19组）可以得知，天涯论坛关于PM2.5的回帖中主要集中于PM2.5新闻数据报道的关注，政府和PM2.5相关性的讨论，以及PM2.5对环境产生影响的讨论。

## 结论

通过数据的挖掘，分析，和主题模型、推荐系统、回帖网络的建立，我们发现天涯杂谈板块并没有形成大规模的关于pm2.5的舆论讨论，热度也主要集中与12、13年，这与pm2.5在网络上的讨论热度集中的年份相符。而具体分析天涯回帖数据的文本和主题，可以发现热议词为“空气”、“口罩”和“污染”，而关注的主题也是pm2.5的影响的数据和怎样面对这种影响。而在推荐系统的建立中，因为部分人只回复了一个帖子，所以存在相似度为零的情况，无法构建推荐系统。天涯杂谈关于PM2.5的回帖网络中，其中心位置，形成了4个小的回帖网络。
